## LAB

```python
blocks = int(input("Enter the number of blocks: "))
height = 0
used_blocks = 0

while blocks:
    diff = blocks - used_blocks
    
    if diff >= (height + 1):
        used_blocks += height + 1
        height += 1
    else:
        break
    

print("The height of the pyramid:", height)
```

```python
c0 = int(input("Enter a positive integer: "))
steps = 0

while c0:
    if c0 < 0:
        print("I said a POSITIVE integer -_-. Try again: ")
    elif c0 == 1:
        break
    elif c0 % 2 == 0:
        c0 /= 2
        steps += 1
        print(c0)
    else:
        c0 = (3 * c0) + 1
        steps += 1
        print(c0)

print("Steps: ", steps)
```

```python
my_list = [1, 2, 4, 4, 1, 4, 2, 6, 2, 9]
```

```python
my_list.sort()
print(my_list)
aux_list = my_list[:]
cond = True

while cond:
    cond = False
    total = range(len(aux_list)-1)
    for i in total:
        if aux_list[i] == aux_list[i + 1]:
            cond = True
            del aux_list[i]
            break
        else:
            continue

print("The list with unique elements only:")
print(aux_list)
```

## LAB

```python
def is_year_leap(year):
    if year % 400 == 0:
        return True
    elif year % 4 == 0 and year % 100 != 0:
        return True
    else:
        return False

def days_in_month(year, month):
    days_m = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    
    if is_year_leap(year) == True:
        if month == 2:
            x = days_m[month - 1]
            return x + 1
        else:
            return days_m[month - 1]
    else:
        return days_m[month - 1]

def day_of_year(year, month, day):
    prev_days = 0
    
    for i in range(1, month):
        prev_days += days_in_month(year, i)
    total = prev_days + day
    return total

print(day_of_year(2000, 12, 31))
```

```python
def is_prime(num):
    count = 0

    if num == 1:
        return False
    elif num == 2:
        return True
    else:
        for i in range(2, num):
            if num % i == 0:
                count += 1
                #print("Count: ", count)
        
        if count == 0:
            return True
        else:
            return False

for i in range(1, 20):
    print(i)
    if is_prime(i + 1):
        print(i + 1, ":", end=" ")
```

```python
def liters_100km_to_miles_gallon(liters):
    ltr_to_gal_100km = liters / 3.785411784
    gal_100_to_1_km = 100 / ltr_to_gal_100km
    km_to_miles = gal_100_to_1_km * (1000 / 1609.344)
    return km_to_miles

def miles_gallon_to_liters_100km(miles):
    miles_to_km = miles * (1609.344 / 1000)
    km_to_100km = 100 / miles_to_km
    return km_to_100km * 3.785411784

print(liters_100km_to_miles_gallon(3.9))
print(liters_100km_to_miles_gallon(7.5))
print(liters_100km_to_miles_gallon(10.))

print(miles_gallon_to_liters_100km(60.3))
print(miles_gallon_to_liters_100km(31.4))
print(miles_gallon_to_liters_100km(23.5))
```

```python
d1 = {'Adam Smith': 'A', 'Judy Paxton': 'B+'}
d2 = {'Mary Louis': 'A', 'Patrick White': 'C'}
d3 = {}

a = (d1, d2)
print(type(a))

for item in (d1, d2):
    for key in item.keys():
        d3[key] = item[key]
print(d3)
```

```python
d1 = {'Adam Smith': 'A', 'Judy Paxton': 'B+'}
d2 = {'Mary Louis': 'A', 'Patrick White': 'C'}
d3 = {}

a = (d1, d2)
print(type(a))

for item in (d1, d2):
    d3.update(item)
print(d3)
```
