```python
def display_board(board):
    # The function accepts one parameter containing the board's current status
    # and prints it out to the console.
    
    print(f"""
    +-------+-------+-------+
    |       |       |       |
    |   {board[0][0]}   |   {board[0][1]}   |   {board[0][2]}   |
    |       |       |       |
    +-------+-------+-------+
    |       |       |       |
    |   {board[1][0]}   |   {board[1][1]}   |   {board[1][2]}   |
    |       |       |       |
    +-------+-------+-------+
    |       |       |       |
    |   {board[2][0]}   |   {board[2][1]}   |   {board[2][2]}   |
    |       |       |       |
    +-------+-------+-------+""")


def enter_move(board):
    # The function accepts the board's current status, asks the user about their move, 
    # checks the input, and updates the board according to the user's decision.
    print("Tu turno (O).", end=" ")
    move = int(input("Haz tu movimiento: "))
    row= positions[move][0]
    col = positions[move][1]

    while True:
        if type(board[row][col]) == int:
            board[row][col] = "O"
            break
        else:
            print("Debes elegir una casilla libre.")
    display_board(board)
    return board


def make_list_of_free_fields(board):
    # The function browses the board and builds a list of all the free squares; 
    # the list consists of tuples, while each tuple is a pair of row and column numbers.

    free = []
    for r in range(len(board)):
        for c in range(len(board[r])):
            if type(board[r][c]) == int:
                free.append((r, c))
            else:
                 continue
    return free
    


def victory_for(board):
    # The function analyzes the board's status in order to check if 
    # the player using 'O's or 'X's has won the game
    for r in range(len(board)): # Revisar filas
        if board[r][0] == board[r][1] and board[r][1] == board[r][2]:
            if board[r][0] == "O":
                print("¡Ganaste!")
                return False
            else:
                print("Ganó la computadora.")
                return False

    for c in range(len(board)):
        if board[0][c] == board[1][c] and board[1][c] == board[2][c]:
            if board[0][c] == "O":
                print("¡Ganaste!")
                return False
            else:
                print("Ganó la computadora.")
                return False

    if board[0][0] == board[1][1] and board[1][1] == board[2][2]:
        if board[0][0] == "O":
                print("¡Ganaste!")
                return False
        else:
                print("Ganó la computadora.")
                return False

    if board[0][2] == board[1][1] and board[1][1] == board[2][0]:
        if board[0][2] == "O":
                print("¡Ganaste!")
                return False
        else:
                print("Ganó la computadora.")
                return False
    return True


def draw_move(board, lst):
    # The function draws the computer's move and updates the board.
    from random import randrange

    print("Turno de la Computadora (X): ")

    while True:
        m_move = randrange(9)
        if positions[m_move] in lst:
            row = positions[m_move][0]
            col = positions[m_move][1]
            board[row][col] = "X"
            break
    display_board(board)
    return board
                


board = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
positions = {1: (0,0), 2: (0,1), 3: (0,2), 4: (1,0), 5: (1,1), 6: (1,2), 7: (2,0), 8:(2,1), 9: (2,2)}

print("""
      --------Juego de Gato--------
      Por @emartlion
      """)
check = True

while check:
    lst = make_list_of_free_fields(board)
    draw_move(board, lst)
    lst = make_list_of_free_fields(board)
    enter_move(board)
    lst = make_list_of_free_fields(board)
    check = victory_for(board)
```
