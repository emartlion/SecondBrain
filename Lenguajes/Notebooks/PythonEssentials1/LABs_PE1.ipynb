```python
def is_year_leap(year):
    if year % 400 == 0:
        return True
    elif year % 4 == 0 and year % 100 != 0:
        return True
    else:
        return False

def days_in_month(year, month):
    days_m = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    
    if is_year_leap(year) == True:
        if month == 2:
            x = days_m[month - 1]
            return x + 1
        else:
            return days_m[month - 1]
    else:
        return days_m[month - 1]

def day_of_year(year, month, day):
    prev_days = 0
    
    for i in range(1, month):
        prev_days += days_in_month(year, i)
    total = prev_days + day
    return total

print(day_of_year(2000, 12, 31))
```

```python
def is_prime(num):
    count = 0

    if num == 1:
        return False
    elif num == 2:
        return True
    else:
        for i in range(2, num):
            if num % i == 0:
                count += 1
                #print("Count: ", count)
        
        if count == 0:
            return True
        else:
            return False

for i in range(1, 20):
    print(i)
    if is_prime(i + 1):
        print(i + 1, ":", end=" ")
```

```python
def liters_100km_to_miles_gallon(liters):
    ltr_to_gal_100km = liters / 3.785411784
    gal_100_to_1_km = 100 / ltr_to_gal_100km
    km_to_miles = gal_100_to_1_km * (1000 / 1609.344)
    return km_to_miles

def miles_gallon_to_liters_100km(miles):
    miles_to_km = miles * (1609.344 / 1000)
    km_to_100km = 100 / miles_to_km
    return km_to_100km * 3.785411784

print(liters_100km_to_miles_gallon(3.9))
print(liters_100km_to_miles_gallon(7.5))
print(liters_100km_to_miles_gallon(10.))

print(miles_gallon_to_liters_100km(60.3))
print(miles_gallon_to_liters_100km(31.4))
print(miles_gallon_to_liters_100km(23.5))
```

```python
def display_board(board):
    # The function accepts one parameter containing the board's current status
    # and prints it out to the console.
    
    print(f"""
    +-------+-------+-------+
    |       |       |       |
    |   {board[0][0]}   |   {board[0][1]}   |   {board[0][2]}   |
    |       |       |       |
    +-------+-------+-------+
    |       |       |       |
    |   {board[1][0]}   |   {board[1][1]}   |   {board[1][2]}   |
    |       |       |       |
    +-------+-------+-------+
    |       |       |       |
    |   {board[2][0]}   |   {board[2][1]}   |   {board[2][2]}   |
    |       |       |       |
    +-------+-------+-------+""")


def enter_move(board):
    # The function accepts the board's current status, asks the user about their move, 
    # checks the input, and updates the board according to the user's decision.
    print("Tu turno (O).", end=" ")
    move = int(input("Haz tu movimiento: "))
    row= positions[move][0]
    col = positions[move][1]

    while True:
        if type(board[row][col]) == int:
            board[row][col] = "O"
            break
        else:
            print("Debes elegir una casilla libre.")
    display_board(board)
    return board


def make_list_of_free_fields(board):
    # The function browses the board and builds a list of all the free squares; 
    # the list consists of tuples, while each tuple is a pair of row and column numbers.

    free = []
    for r in range(len(board)):
        for c in range(len(board[r])):
            if type(board[r][c]) == int:
                free.append((r, c))
            else:
                 continue
    return free
    


def victory_for(board):
    # The function analyzes the board's status in order to check if 
    # the player using 'O's or 'X's has won the game
    for r in range(len(board)): # Revisar filas
        if board[r][0] == board[r][1] and board[r][1] == board[r][2]:
            if board[r][0] == "O":
                print("¡Ganaste!")
                return False
            else:
                print("Ganó la computadora.")
                return False

    for c in range(len(board)):
        if board[0][c] == board[1][c] and board[1][c] == board[2][c]:
            if board[0][c] == "O":
                print("¡Ganaste!")
                return False
            else:
                print("Ganó la computadora.")
                return False

    if board[0][0] == board[1][1] and board[1][1] == board[2][2]:
        if board[0][0] == "O":
                print("¡Ganaste!")
                return False
        else:
                print("Ganó la computadora.")
                return False

    if board[0][2] == board[1][1] and board[1][1] == board[2][0]:
        if board[0][2] == "O":
                print("¡Ganaste!")
                return False
        else:
                print("Ganó la computadora.")
                return False
    return True


def draw_move(board, lst):
    # The function draws the computer's move and updates the board.
    from random import randrange

    print("Turno de la Computadora (X): ")

    while True:
        m_move = randrange(9)
        if positions[m_move] in lst:
            row = positions[m_move][0]
            col = positions[m_move][1]
            board[row][col] = "X"
            break
    display_board(board)
    return board
                


board = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
positions = {1: (0,0), 2: (0,1), 3: (0,2), 4: (1,0), 5: (1,1), 6: (1,2), 7: (2,0), 8:(2,1), 9: (2,2)}

print("""
      --------Juego de Gato--------
      Por @emartlion
      """)
check = True

while check:
    lst = make_list_of_free_fields(board)
    draw_move(board, lst)
    lst = make_list_of_free_fields(board)
    enter_move(board)
    lst = make_list_of_free_fields(board)
    check = victory_for(board)
```
